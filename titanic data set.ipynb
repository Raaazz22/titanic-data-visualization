Visualize the titanic data set 

```python
## import the seaborn library and matplotlib.pyplot

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load the titanic dataset using the seaborn library

titanic_data = sns.load_dataset('titanic')
```

```python
#dataset

titanic_data
```

```python
#show the head of data

titanic_data.head(5)
```

```python
# hiost plot the gender data using matplotlib

plt.hist(x = 'sex' , data = titanic_data)
```

```python
# hiost plot the pclass data using matplotlib

plt.hist(x = 'pclass' , data = titanic_data )
```

```python
# hiost plot the alive data using matplotlib

plt.hist(x = 'alive' , data = titanic_data)
```

```python
#histogram plot using the seaborn

sns.displot(x = 'pclass' , data = titanic_data , height =2)
```

```python
#categorical plot using seaborn

sns.catplot(x = 'sex' , y = 'who' , data = titanic_data , height =3)
```

```python
import seaborn as sns
# For Plot 1
sns.jointplot(x = titanic_data['age'], y = titanic_data['fare'], kind = 'scatter')
# For Plot 2
sns.jointplot(x = titanic_data['age'], y = titanic_data['fare'], kind = 'hex')
```

```python
#bar plot using seborn

import seaborn as sns
sns.barplot(x = 'sex', y = 'fare' , data = titanic_data)
```
